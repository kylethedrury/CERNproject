```python
# IMPORTING PACKAGES

import numpy as np                                # import packages and give them quick handles since they will be used often
import matplotlib.pyplot as plt
from scipy import optimize
import pandas as pd
import math as math
```

```python
# gg2gGamma OR gq2qGamma? 1 means gg, 0 means gq
# Specify Ms and COME (TeV)

Ms = 6000
flag = 1
COME = 13000
```

```python
# READ LHE LINES

lhe_file = open('sample.lhe', 'r')
lines = lhe_file.readlines()
# print(lines)
lhe_file.close()
```

```python
# Initialize arrays for all relevant data

# INITIAL PARTONS ... do we need these?

g1i_xMomentum, g1i_yMomentum, g1i_zMomentum, g1i_Energy = [], [], [], [] # 4-vector of first initial gluon
g2i_xMomentum, g2i_yMomentum, g2i_zMomentum, g2i_Energy = [], [], [], [] # 4-vector of second initial gluon
qi_xMomentum, qi_yMomentum, qi_zMomentum, qi_Energy = [], [], [], [] # 4-vector of first initial quark

# FINAL PARTONS

gf_xMomentum, gf_yMomentum, gf_zMomentum, gf_Energy = [], [], [], [] # 4-vector of final gluon
qf_xMomentum, qf_yMomentum, qf_zMomentum, qf_Energy = [], [], [], [] # 4-vector of final quark
Gamma_xMomentum, Gamma_yMomentum, Gamma_zMomentum, Gamma_Energy = [], [], [], [] # 4-vector of photon

# 3-MOMENTA OF FINAL PARTONS, s-hat for each collison, and invariant mass

Gamma_Momentum, g_Momentum, q_momentum = [], [], []
s_hat, invariant_Mass = [], []

# RAPIDITY STUFF
y_Gamma, y_gf, y_qf, y, Y, p_perp = [], [], [], [], [], []
```

```python
# PARSE TIME GO BRRR

i = 0

if flag == 1:                                     # gg2gGamma case

    for line in enumerate(lines):
    
        if line[0:17] == '21 -1 0 0 103 101':     
            g1i_xMomentum.append(line[19:31])
            g1i_yMomentum.append(line[33:45])
            g1i_zMomentum.append(line[47:59])
            g1i_Energy.append(line[61:73])
            print()
        
        if line[0:17] == '21 -1 0 0 101 102':      
            g2i_xMomentum.append(line[19:31])
            g2i_yMomentum.append(line[33:45])
            g2i_zMomentum.append(line[47:59])
            g2i_Energy.append(line[61:73])
            
        if line[0:17] == '21  1 1 2 103 102':     
            gf_xMomentum.append(line[19:31])
            gf_yMomentum.append(line[33:45])
            gf_zMomentum.append(line[47:59])
            gf_Energy.append(line[61:73])
            
        if line [0:17] == '22  1 1 2   0   0':
            Gamma_xMomentum.append(line[19:31])
            Gamma_yMomentum.append(line[33:45])
            Gamma_zMomentum.append(line[47:59])
            Gamma_Energy.append(line[61:73])
        
        
if flag == 0:                                      # gq2qGamma case
    
    for line in enumerate(lines):
        
        if line[0:17] == ' 1 -1 0 0 101   0':      
            qi_xMomentum.append(line[19:31])
            qi_yMomentum.append(line[33:45])
            qi_zMomentum.append(line[47:59])
            qi_Energy.append(line[61:73])
            
        if line[0:17] == '21 -1 0 0 103 101':     
            g1i_xMomentum.append(line[19:31])
            g1i_yMomentum.append(line[33:45])
            g1i_zMomentum.append(line[47:59])
            g1i_Energy.append(line[61:73])
            print()
         
        if line[0:17] == ' 1  1 1 2 103   0':      
            qf_xMomentum.append(line[19:31])
            qf_yMomentum.append(line[33:45])
            qf_zMomentum.append(line[47:59])
            qf_Energy.append(line[61:73])
        
        if line [0:17] == '22  1 1 2   0   0':
            Gamma_xMomentum.append(line[19:31])
            Gamma_yMomentum.append(line[33:45])
            Gamma_zMomentum.append(line[47:59])
            Gamma_Energy.append(line[61:73])
    
    i = i+1     # counter
```

```python
# Convert string arrays into float arrays

if flag == 1:                                            # gg2gGamma case

    for i in range(len(g1i_Energy)):
        g1i_xMomentum[i] = float(g1i_xMomentum[i])
        g1i_yMomentum[i] = float(g1i_yMomentum[i])
        g1i_zMomentum[i] = float(g1i_zMomentum[i])
        g1i_Energy[i] = float(g1i_Energy[i])
    
        g2i_xMomentum[i] = float(g2i_xMomentum[i])     
        g2i_yMomentum[i] = float(g2i_yMomentum[i])
        g2i_zMomentum[i] = float(g2i_zMomentum[i])
        g2i_Energy[i] = float(g2i_Energy[i])
        
        gf_xMomentum[i] = float(gf_xMomentum[i])         
        gf_yMomentum[i] = float(gf_yMomentum[i])
        gf_zMomentum[i] = float(gf_zMomentum[i])
        gf_Energy[i] = float(gf_Energy[i])
        
        Gamma_xMomentum[i] = float(Gamma_xMomentum[i])
        Gamma_yMomentum[i] = float(Gamma_yMomentum[i])
        Gamma_zMomentum[i] = float(Gamma_zMomentum[i])
        Gamma_Energy[i] = float(Gamma_Energy[i])
 

if flag == 0:                                             # gq2qGamma case

    for i in range(len(g1i_Energy)):
        g1i_xMomentum[i] = float(g1i_xMomentum[i])
        g1i_yMomentum[i] = float(g1i_yMomentum[i])
        g1i_zMomentum[i] = float(g1i_zMomentum[i])
        g1i_Energy[i] = float(g1i_Energy[i])
        
        qi_xMomentum[i] = float(qi_xMomentum[i])         
        qi_yMomentum[i] = float(qi_yMomentum[i])
        qi_zMomentum[i] = float(qi_zMomentum[i])
        qi_Energy[i] = float(qi_Energy[i])

        qf_xMomentum[i] = float(qf_xMomentum[i])        
        qf_yMomentum[i] = float(qf_yMomentum[i])
        qf_zMomentum[i] = float(qf_zMomentum[i])
        qf_Energy[i] = float(qf_Energy[i])
    
        Gamma_xMomentum[i] = float(Gamma_xMomentum[i])
        Gamma_yMomentum[i] = float(Gamma_yMomentum[i])
        Gamma_zMomentum[i] = float(Gamma_zMomentum[i])
        Gamma_Energy[i] = float(Gamma_Energy[i])
```

```python
# 3-MOMENTA, S_HAT, INVARIANT MASS of FINAL PARTONS, RAPIDITIES

if flag == 1:                               # gg2gGamma case

    for i in range(len(g1i_Energy)):
    
        Gamma_Momentum = sqrt( Gamma_Momentum[i]**2 + Gamma_yMomentum[i]**2 + Gamma_zMomentum[i]**2 )
        gf_Momentum = sqrt( gf_Momentum[i]**2 + gf_yMomentum[i]**2 + gf_zMomentum[i]**2 )

        s_hat.append( (Gamma_Momentum + gf_Momentum)**2 )
    
        invariant_Mass.append( sqrt(s_hat[i]) )
        
        y_Gamma.append( 0.5 * ln( (Gamma_Energy[i] + Gamma_zMomentum) / (Gamma_Energy[i] - Gamma_zMomentum) ) )
        y_gf.append( 0.5 * ln( (gf_Energy[i] + gf_zMomentum) / (gf_Energy[i] - gf_zMomentum) ) )
        
        y.append( abs( 0.5*(y_Gamma[i] - y_gf[i]) ) )
        Y.append( abs( 0.5*(y_Gamma[i] + y_gf[i]) ) )
        p_perp.append( invariant_Mass / (2*cosh(y[i])) )

        
if flag == 0:                                # gq2qGamma case

    for i in range(len(g1i_Energy)):
    
        Gamma_Momentum = sqrt( Gamma_Momentum[i]**2 + Gamma_yMomentum[i]**2 + Gamma_zMomentum[i]**2 )
        qf_Momentum = sqrt( qf_Momentum[i]**2 + qf_yMomentum[i]**2 + qf_zMomentum[i]**2 )

        s_hat.append( (Gamma_Momentum + qf_Momentum)**2 )
    
        invariant_Mass.append( sqrt(s_hat[i]) )
        
        y_Gamma.append( 0.5 * ln( (Gamma_Energy[i] + Gamma_zMomentum) / (Gamma_Energy[i] - Gamma_zMomentum) ) )
        y_qf.append( 0.5 * ln( (gf_Energy[i] + gf_zMomentum) / (gf_Energy[i] - gf_zMomentum) ) )
        
        y.append( abs( 0.5*(y_Gamma[i] - y_qf[i]) ) )
        Y.append( abs( 0.5*(y_Gamma[i] + y_qf[i]) ) )
        p_perp.append( invariant_Mass / (2*math.cosh(y[i])) )
```

```python
# PREPARE TO CALCULATE THE DIFFERENTIAL CROSS-SECTIONS OF EACH EVENT

# String coupling g and alpha at Ms

alpha = (1/0.118) - ( 7/(2*math.pi) * math.log(Ms/91.2) )  # units of GeV
g = math.sqrt(4*math.pi*alpha)

# FIRST RESONANCE WIDTHS

GJ0_g = 3 * (g**2) * Ms / (16*math.pi)
GJ2_g = 9 * (g**2) * Ms / (80*math.pi)
GJ12_q = 3 * (g**2) * Ms / (32*math.pi)
GJ32_q = 3 * (g**2) * Ms / (64*math.pi)

# GET Q

kappa = 0.1414
sin2ThetaW = 0.23121
cosThetaW = math.sqrt(1 - sin2ThetaW)
Q = math.sqrt(1/6) * kappa * cosThetaW

# Calculate all of our t_hats and u_hats
t_hat, u_hat = [], []

for i in range(len(s_hat)):
    t_hat.append( -(invariant_Mass[i]**2/2) * exp(-y[i]) / math.cosh(y[i]) )
    u_hat.append( -(invariant_Mass[i]**2/2) * exp(-y[i]) / math.cosh(y[i]) )

# STRING AMPLITUDES Msq for gg2Gamma and gq2qGamma

Msq = []

if flag == 1:     # gg2gGamma case
    
    for i in range(len(s_hat)):
        
        amplitude = 5 * g**4 * Q**2 / 3 / Ms**4 \
        * ( Ms**8 / ( (s_hat[i]-Ms**2)**2 + (GJ0_g*Ms)**2 ) \
        + ( (t_hat[i]**4 + u_hat[i]**4) / ( (s_hat[i]-Ms**2)**2 + (GJ2_g*Ms)**2 ) ) )
        
        Msq.append(amplitude)
        
if flag == 0:     #gq2qGamma case
    
    for i in range(len(s_hat)):
        
        amplitude = - g**4 * Q**2 / 3 / Ms**2 \
        * ( Ms**4 / ( (s_hat[i]-Ms**2)**2 + (GJ12_q*Ms)**2 ) \
        + ( (u_hat[i]**3) / ( (s_hat[i]-Ms**2)**2 + (GJ32_g*Ms)**2 ) ) )
        
        Msq.append(amplitude)
        
# Let's get tauuuu

tau = []

for i in range(len(invariant_Mass)):
    tau.append(invariant_Mass[i] / COME)
    
    
# x_a, x_b y_max and Y_max

x_a, x_b = [], []
y_max, Y_max = [], []

for i in range(len(tau)):

    x_a.append(math.sqrt(tau)*exp(Y[i]))
    x_b.append(math.sqrt(tau)*exp(-Y[i]))
    
    if flag == 1:                           # gg2gGamma case
        y_max.append( max(y_gf, y_Gamma) )
        
    if flag == 0:                           # gq2qGamma case
        y_max.append( max(y_qf, y_Gamma) )
        
    Y_max.append( min(math.log(tau[i]**(-1/2), y_max)) )
```
